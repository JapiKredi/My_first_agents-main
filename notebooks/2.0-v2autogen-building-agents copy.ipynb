# AutoGen Basics
## 4 Basic steps
1. Define configuration list
2. Define the assistant
3. Define the user
4. Initiate chat (solve the task)

```python
%matplotlib inline
import autogen
from autogen import AssistantAgent, UserProxyAgent
from autogen import config_list_from_json
from autogen.coding import LocalCommandLineCodeExecutor
```

```python
import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv('/Users/jasper/Downloads/My_first_agents-main/notebooks/.env')

# Verify that the API key is loaded
api_key = os.getenv('OPENAI_API_KEY')
```

```python
config_list = [
    {"model": "gpt-3.5-turbo-16k", 
              "api_key": os.environ["OPENAI_API_KEY"]}
]

llm_config={
    "timeout": 600,
    "seed": 42,
    "config_list": config_list,
    "temperature": 0
}
```

```python
assistant = autogen.AssistantAgent(
    name="assistant", 
    llm_config=llm_config,
    system_message="You are an assistant agent. You can ask for help with any task or problem you have."
    )

user_proxy = autogen.UserProxyAgent(
    name="user_proxy", 
    human_input_mode="TERMINATE",
    max_consecutive_auto_reply=10,
    is_termination_msg=lambda x: x.get("content","").rstrip().endswith("TERMINATE"),
    code_execution_config={"executor": autogen.coding.LocalCommandLineCodeExecutor(work_dir="web")},
    #code_execution_config={"executor": autogen.coding.LocalCommandLineCodeExecutor(work_dir="coding-21dec")},
    llm_config = llm_config,
    system_message = """Reply with TERMINATE if the task has been solved at full satisfaction. otherwise reply CONTINUE or the reason why the task has not been solved yet."""
)

task = """
Plot a chart of NVDA and TESLA stock price change YTD and save the result in stock_chart.png .
"""
```

```python
# Start the chat
user_proxy.initiate_chat(
    assistant,
    messages=task,
)
```
